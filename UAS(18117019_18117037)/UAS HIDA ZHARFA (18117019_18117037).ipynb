```python
import pandas as pd
import numpy as np
import csv
```

```python
#import csv DATASISWA sebagai variabel article_read
#berisi nama lengkap, asal SMA, Pil 1 2 3 Universitas dan Jurusan, serta nilai UTBK 1

article_read = pd.read_csv('DATASISWA.csv', delimiter = ',', names = ['NAMA_LENGKAP', 'ASAL_SMA', 'PILIHAN_1_UNIVERSITAS', 'PILIHAN_1_JURUSAN', 'PILIHAN_2_UNIVERSITAS', 'PILIHAN_2_JURUSAN', 'PILIHAN_3_UNIVERSITAS', 'PILIHAN_3_JURUSAN', 'NILAI_UTBK_1_TKA', 'NILAI_UTBK_1_TPS', 'NILAI_UTBK_2_TKA', 'NILAI_UTBK_2_TPS'])
article_read 
```

```python
#import csv DATAKAMPUS sebagai variabel article_read2
#berisi nama kampus, jurusan, daya tampung, passing grade, nominal UKT dari gelombang 1 hingga 8

article_read2 = pd.read_csv('DATAKAMPUS.csv', delimiter = ',', names = ['NAMA_KAMPUS', 'JURUSAN', 'DAYA_TAMPUNG', 'PASSING_GRADE', 'GEL_1', 'GEL_2', 'GEL_3', 'GEL_4', 'GEL_5', 'GEL_6', 'GEL_7', 'GEL_8'])
article_read2
```

```python
#Membuat data frame dengan variabel INPUTAN
#yang berisi contoh inputan siswa berupa nama, asal SMA, pilihan 1 2 3 Universitas dan jurusan, serta nilai UTBK (TPS dan TKA) 1 & 2
#dengan urutan disesuaikan database di atas

INPUTAN = pd.DataFrame([['ZHARFA','SMA SEMESTA','UNIVERSITAS INDONESIA','TEKNIK ELEKTRO','UNIVERSITAS GAJAH MADA','KEDOKTERAN','UNIVERSITAS GAJAH MADA','TEKNIK ELEKTRO', '700', '650', '830', '650']], columns=['NAMA_LENGKAP', 'ASAL_SMA', 'PILIHAN_1_UNIVERSITAS', 'PILIHAN_1_JURUSAN', 'PILIHAN_2_UNIVERSITAS', 'PILIHAN_2_JURUSAN', 'PILIHAN_3_UNIVERSITAS', 'PILIHAN_3_JURUSAN', 'NILAI_UTBK_1_TKA', 'NILAI_UTBK_1_TPS', 'NILAI_UTBK_2_TKA', 'NILAI_UTBK_2_TPS'])
INPUTAN
```

```python
#Menggabungkan Inputan siswa dengan database siswa dengan variabel hasil

hasil = pd.concat([article_read, INPUTAN], ignore_index=True, sort=False)
hasil
```

```python
#Menampilkan data "hasil" hanya untuk yang pilihan 1 universitasnya merupakan Universitas Indonesia
#Disimpan dengan variabel UNI1

UNI1=hasil[hasil.PILIHAN_1_UNIVERSITAS == 'UNIVERSITAS INDONESIA']
UNI1
```

```python
#Menampilkan data "UNI1" hanya untuk yang pilihan 1 jurusannya merupakan Teknik Elektro
#Disimpan dengan variabel JURUSAN1
#Dengan kata lain, menampilkan data "hasil" hanya untuk yang pilihan 1 nya adalah Teknik Elektro-Universitas Indonesia

JURUSAN1 = UNI1[UNI1.PILIHAN_1_JURUSAN == 'TEKNIK ELEKTRO']
JURUSAN1
```

```python
#Menampilkan data "hasil" hanya untuk yang pilihan 2 universitasnya merupakan Universitas Indonesia
#Disimpan dengan variabel UNI2

UNI2=hasil[hasil.PILIHAN_2_UNIVERSITAS == 'UNIVERSITAS INDONESIA']
UNI2
```

```python
#Menampilkan data "UNI2" hanya untuk yang pilihan 2 jurusannya merupakan Teknik Elektro
#Disimpan dengan variabel JURUSAN2
#Dengan kata lain, menampilkan data "hasil" hanya untuk yang pilihan 2 nya adalah Teknik Elektro-Universitas Indonesia

JURUSAN2 = UNI2[UNI2.PILIHAN_2_JURUSAN == 'TEKNIK ELEKTRO']
JURUSAN2
```

```python
#Menampilkan data "hasil" hanya untuk yang pilihan 3 universitasnya merupakan Universitas Indonesia
#Disimpan dengan variabel UNI3

UNI3=hasil[hasil.PILIHAN_3_UNIVERSITAS == 'UNIVERSITAS INDONESIA']
UNI3
```

```python
#Menampilkan data "UNI3" hanya untuk yang pilihan 3 jurusannya merupakan Teknik Elektro
#Disimpan dengan variabel JURUSAN3
#Dengan kata lain, menampilkan data "hasil" hanya untuk yang pilihan 3 nya Teknik Elektro-Universitas Indonesia

JURUSAN3 = UNI3[UNI3.PILIHAN_3_JURUSAN == 'TEKNIK ELEKTRO']
JURUSAN3
```

```python
#Menggabungkan seluruh data UNI1, UNI2, UNI3 dengan variabel semua
#Dengan kata lain, data "semua" berisi data siswa dengan pilihan 1, 2, atau 3 nya merupakan Teknik Elektro-Universitas Indonesia

semua = pd.concat([JURUSAN1,JURUSAN2,JURUSAN3], sort=False)
semua
```

```python
#Menampilkan kolom nilai UTBK 1 dan 2 dari data "JURUSAN1" kemudian di-transpose
#Disimpan dengan variabel FILTERrata

FILTERrata = JURUSAN1[['NILAI_UTBK_1_TKA', 'NILAI_UTBK_1_TPS', 'NILAI_UTBK_2_TKA', 'NILAI_UTBK_2_TPS']].astype(int).T
FILTERrata
```

```python
#Menghitung nilai rata-rata dari nilai UTBK tiap siswa pada data "FILTERrata"
#mengubah hasilnya ke dalam csv, disimpan dengan variabel rata
#Import data csv "rata" dengan variabel RATA

rata = FILTERrata.mean().to_csv('rata.csv')
RATA = pd.read_csv('rata.csv', delimiter = ',', names = ['RATA_RATA_UTBK'])
RATA
```

```python
#Menggabungkan data "JURUSAN1" yang ditranspose dengan data "RATA" yang ditranspose, disimpan dengan variabel gabungan1
#Data "gabungan1" ditranspose kemudian dilakukan sorting berdasarkan nilai rata-rata UTBK dari yang tertinggi ke terendah
#Hasilnya disimpan dengan variabel GABUNGAN1

gabungan1 = pd.concat([JURUSAN1.T, RATA.T], sort=False)
GABUNGAN1 = gabungan1.T.sort_values(by =['RATA_RATA_UTBK'], ascending=False)
GABUNGAN1
```

```python
#Menampilkan kolom nilai UTBK 1 dan 2 dari data "JURUSAN2" kemudian di-transpose
#Disimpan dengan variabel FILTERrata2

FILTERrata2 = JURUSAN2[['NILAI_UTBK_1_TKA', 'NILAI_UTBK_1_TPS', 'NILAI_UTBK_2_TKA', 'NILAI_UTBK_2_TPS']].astype(int).T
FILTERrata2
```

```python
#Menghitung nilai rata-rata dari nilai UTBK tiap siswa pada data "FILTERrata2"
#mengubah hasilnya ke dalam csv, disimpan dengan variabel rata2
#Import data csv "rata2" dengan variabel RATA2

rata2 = FILTERrata2.mean().to_csv('rata2.csv')
RATA2 = pd.read_csv('rata2.csv', delimiter = ',', names = ['RATA_RATA_UTBK'])
RATA2
```

```python
#Menggabungkan data "JURUSAN2" yang ditranspose dengan data "RATA2" yang ditranspose, disimpan dengan variabel gabungan2
#Data "gabungan2" ditranspose kemudian dilakukan sorting berdasarkan nilai rata-rata UTBK dari yang tertinggi ke terendah
#Hasilnya disimpan dengan variabel GABUNGAN2

gabungan2 = pd.concat([JURUSAN2.T, RATA2.T], sort=False, )
GABUNGAN2 = gabungan2.T.sort_values(by =['RATA_RATA_UTBK'], ascending=False)
GABUNGAN2
```

```python
#Menampilkan kolom nilai UTBK 1 dan 2 dari data "JURUSAN3" kemudian di-transpose
#Disimpan dengan variabel FILTERrata3

FILTERrata3 = JURUSAN3[['NILAI_UTBK_1_TKA', 'NILAI_UTBK_1_TPS', 'NILAI_UTBK_2_TKA', 'NILAI_UTBK_2_TPS']].astype(int).T
FILTERrata3
```

```python
#Menghitung nilai rata-rata dari nilai UTBK tiap siswa pada data "FILTERrata3"
#mengubah hasilnya ke dalam csv, disimpan dengan variabel rata3
#Import data csv "rata3" dengan variabel RATA3

rata3 = FILTERrata3.mean().to_csv('rata3.csv')
RATA3 = pd.read_csv('rata3.csv', delimiter = ',', names = ['RATA_RATA_UTBK'])
RATA3
```

```python
#Menggabungkan data "JURUSAN3" yang ditranspose dengan data "RATA3" yang ditranspose, disimpan dengan variabel gabungan3
#Data "gabungan3" ditranspose kemudian dilakukan sorting berdasarkan nilai rata-rata UTBK dari yang tertinggi ke terendah
#Hasilnya disimpan dengan variabel GABUNGAN3

gabungan3 = pd.concat([JURUSAN3.T, RATA3.T], sort=False, )
GABUNGAN3 = gabungan3.T.sort_values(by =['RATA_RATA_UTBK'], ascending=False)
GABUNGAN3
```

```python
#Menggabungkan data GABUNGAN1, GABUNGAN2, dan GABUNGAN3 dengan variabel utbk
#Dengan kata lain, utbk berisi data siswa dengan nilai rata-rata utbk yang sudah terurut dan berdasarkan prioritas pilihan jurusan tertinggi
#Dilakukan pengubahan indeks mulai dari 1 untuk menampilkan peringkat siswa

utbk = pd.concat([GABUNGAN1, GABUNGAN2, GABUNGAN3], sort=False, ignore_index=True)
utbk.index += 1
utbk
```
